1. Write a program that accepts a list from user and print the alternate element of list. 


```python
# Taking list input from user
lst = input("Enter elements separated by space: ").split()

# Printing alternate elements
print("Alternate elements are:")
for i in range(0, len(lst), 2):
    print(lst[i])
```

2. Write a program that accepts a list from user. Your program should reverse the content of list and
display it. Do not use reverse() method.

```python
# Taking list input from user
lst = input("Enter elements separated by space: ").split()

# Reversing list manually
reversed_list = []

for i in range(len(lst)-1, -1, -1):
    reversed_list.append(lst[i])

# Display result
print("Reversed list is:", reversed_list)
```

3. Find and display the largest number of a list without using built-in function max(). Your program
should ask the user to input values in list from keyboard. 


```python
# Taking input from user
lst = list(map(int, input("Enter numbers separated by space: ").split()))

# Assume first element is largest
largest = lst[0]

# Loop through list
for num in lst:
    if num > largest:
        largest = num

# Display result
print("Largest number is:", largest)
```

 4. Write a program that rotates the element of a list so that the element at the first index moves to the
second index, the element in the second index moves to the third index, etc., and the element in the last
index moves to the first index. 


```python
# Taking input from user
lst = input("Enter elements separated by space: ").split()

# Rotate list
last = lst[-1]   # store last element

# Shift elements to the right
for i in range(len(lst)-1, 0, -1):
    lst[i] = lst[i-1]

# Put last element at first position
lst[0] = last

# Display result
print("Rotated list is:", lst)
```

5. Write a program that input a string and ask user to delete a given word from a string. 


```python
# Taking string input
text = input("Enter a sentence: ")

# Word to delete
word = input("Enter word to delete: ")

# Split sentence into words
words = text.split()

# Remove the word
new_words = []
for w in words:
    if w != word:
        new_words.append(w)

# Join back into string
result = " ".join(new_words)

# Display result
print("Updated string:", result)
```

 6. Write a program that reads a string from the user containing a date in the form mm/dd/yyyy. It
should print the date in the form March 12, 2021.

```python
# Taking input from user
date = input("Enter date (mm/dd/yyyy): ")

# Split the date
parts = date.split("/")

month = int(parts[0])
day = parts[1]
year = parts[2]

# List of month names
months = ["January", "February", "March", "April", "May", "June",
          "July", "August", "September", "October", "November", "December"]

# Convert and display
print(months[month - 1], day + ",", year)
```

 7. Write a program with a function that accepts a string from keyboard and create a new string after
converting character of each word capitalized. For instance, if the sentence is "stop and smell the roses."
the output should be "Stop And Smell The Roses" 


```python
# Function to capitalize each word
def capitalize_words(text):
    words = text.split()      # split into words
    new_words = []

    for w in words:
        new_word = w[0].upper() + w[1:]   # capitalize first letter
        new_words.append(new_word)

    return " ".join(new_words)   # join back into string


# Taking input from user
sentence = input("Enter a sentence: ")

# Calling function
result = capitalize_words(sentence)

# Display result
print("Output:", result)
```

8. Find the sum of each row of matrix of size m x n. For example for the following matrix output will be
like this :
Sum of row 1 = 32
Sum of row 2 = 31
Sum of row 3 = 63 


```python
# Taking matrix size
m = int(input("Enter number of rows: "))
n = int(input("Enter number of columns: "))

matrix = []

# Taking matrix input
print("Enter matrix values row by row:")
for i in range(m):
    row = list(map(int, input().split()))
    matrix.append(row)

# Calculating sum of each row
for i in range(m):
    row_sum = 0
    for j in range(n):
        row_sum += matrix[i][j]
    
    print(f"Sum of row {i+1} = {row_sum}")
```

9. Write a program to add two matrices of size n x m. 


```python
# Taking matrix size
n = int(input("Enter number of rows: "))
m = int(input("Enter number of columns: "))

# First matrix
print("Enter elements of first matrix:")
A = []
for i in range(n):
    row = list(map(int, input().split()))
    A.append(row)

# Second matrix
print("Enter elements of second matrix:")
B = []
for i in range(n):
    row = list(map(int, input().split()))
    B.append(row)

# Result matrix
result = []

# Adding matrices
for i in range(n):
    row = []
    for j in range(m):
        row.append(A[i][j] + B[i][j])
    result.append(row)

# Display result
print("Sum of matrices is:")
for row in result:
    print(row)
```

10. Write a program to multiply two matrices 

```python
# Taking size of matrices
n = int(input("Enter rows of first matrix: "))
m = int(input("Enter columns of first matrix / rows of second: "))
p = int(input("Enter columns of second matrix: "))

# First matrix A (n x m)
print("Enter elements of first matrix:")
A = []
for i in range(n):
    row = list(map(int, input().split()))
    A.append(row)

# Second matrix B (m x p)
print("Enter elements of second matrix:")
B = []
for i in range(m):
    row = list(map(int, input().split()))
    B.append(row)

# Result matrix (n x p) initialized with 0
result = []
for i in range(n):
    row = [0] * p
    result.append(row)

# Matrix multiplication logic
for i in range(n):
    for j in range(p):
        for k in range(m):
            result[i][j] += A[i][k] * B[k][j]

# Display result
print("Multiplication result is:")
for row in result:
    print(row)
```
